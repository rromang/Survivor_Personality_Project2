```python
# Read in data 
data_file = r"../castaways.csv"
castaways_df = pd.read_csv(data_file)
castaways_df
```

```python
# Use geopy to add latitude and longitude by iterrating through city and state

# Create Blank columns for Lat and Lon
castaways_df["Lat"] = ""
castaways_df["Lon"] = ""


for index, row in castaways_df.iterrows():
    # define city and state parameters
    city = row['city']
    state = row['state']
    location_string = f'{city}, {state}'
    
    # Use geolocator to get lat and lon values from "city, state"
    geolocator = Nominatim(user_agent=g_key)
    location = geolocator.geocode(location_string)
    lat = location.latitude
    lon = location.longitude
    castaways_df.loc[index, 'Lat'] = lat
    castaways_df.loc[index, 'Lon'] = lon

castaways_df.head()
```
