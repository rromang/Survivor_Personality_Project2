```python
import pandas as pd
import numpy as np
```

```python
csv_file = r'../castaways.csv'
castaways_df = pd.read_csv(csv_file)
castaways_df
```

```python
# THIS IS THE MAIN CODE FOR ADDING THE NEW COLUMN AND PERSONALITY NAMES

conditions = [
    (castaways_df['personality_type'] == "INTJ"), # Architect
    (castaways_df['personality_type'] == "INTP"), # Logician
    (castaways_df['personality_type'] == "ENTJ"), # Commander
    (castaways_df['personality_type'] == "ENTP"), # Debater
    (castaways_df['personality_type'] == "INFJ"), # Advocate
    (castaways_df['personality_type'] == "INFP"), # Mediator
    (castaways_df['personality_type'] == "ENFJ"), # Protagonist
    (castaways_df['personality_type'] == "ENFP"), # Campaigner
    (castaways_df['personality_type'] == "ISTJ"), # Logistician
    (castaways_df['personality_type'] == "ISFJ"), # Defender
    (castaways_df['personality_type'] == "ESTJ"), # Executive
    (castaways_df['personality_type'] == "ESFJ"), # Consul
    (castaways_df['personality_type'] == "ISTP"), # Virtuoso
    (castaways_df['personality_type'] == "ISFP"), # Adventurer
    (castaways_df['personality_type'] == "ESTP"), # Entrepreneur
    (castaways_df['personality_type'] == "ESFP") # Entertainer
]

values = ["Architect", "Logician", "Commander", "Debater", "Advocate", "Mediator", "Protagonist", "Campaigner", "Logistician", "Defender", "Executive", "Consul", "Virtuoso", "Adventurer", "Entrepreneur", "Entertainer"]

castaways_df["personality_name"] = np.select(conditions, values)

castaways_df.head()
```
